```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
world_bank = pd.read_excel("WorldBank.xlsx").rename({"Country Name": "Country"}, axis = 1)

world_bank["Population (M)"] = (world_bank["GDP (USD)"] / world_bank["GDP per capita (USD)"] / 1000000)

world_bank.head()
```

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi = pd.read_csv("HDI.csv")

hdi.head()
```

```python
# Left join for only year 2014

hdi_world_2014 = world_bank.query("Year == 2014").merge(hdi[["iso3", "hdi_2014"]], how = "left", 
                                                              left_on = "Country Code", right_on = "iso3")

hdi_world_2014.head()
```

```python
hdi_world_2014.describe()
```

```python
correlation = hdi_world_2014['GDP per capita (USD)'].corr(hdi_world_2014['hdi_2014'])
```

```python
correlation
```

```python
# How GDP per capita & hdi are correlated

sns.lmplot(x = "GDP per capita (USD)",
           y = "hdi_2014",
           data = hdi_world_2014,
          )
```

```python
# Prepare data for later visualizations

# Pivot table: Year to Row, Region to Columns, GDP values and sum agg

gdp_pivot = world_bank.pivot_table(index = "Year", columns = "Region", values = "GDP (USD)", aggfunc = "sum")

gdp_pivot.head()
```

```python
# Pivot table: Year to Row, Region to Columns, Population values and sum agg

pop_pivot = world_bank.pivot_table(index = "Year", columns = "Region", values = "Population (M)", aggfunc = "sum")

pop_pivot.head()
```

```python
# HDI table group by Region, hdi_2014 agg by mean from high to low

wb_hdi_by_region = (hdi_world_2014
                    .groupby("Region")
                    .agg({"hdi_2014": "mean"})
                    .sort_values("hdi_2014", ascending = False)
                   )
wb_hdi_by_region
```

```python
# Staked area chart showing growth of GDP over time, region wise. Sorted by region values
sns.set_style("white")
fig, ax = plt.subplots(figsize= (10,8))

ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region]/1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending= False).index],
    labels = gdp_pivot.iloc[-1].sort_values(ascending= False).index,
    
)
ax.legend(loc = "upper left")
ax.set_title("Growth of GDP over Time")
ax.set_xlabel("Year")
ax.set_xticks(gdp_pivot.index[::5])
ax.set_ylabel("GDP (Trillions) ")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False);
```

```python
gdp_pivot.iloc[-1].sort_values(ascending= False).index
```

```python
# Staked area chart showing growth of Population over time, region wise. Sorted by region values
#sns.set_style("white")
fig, ax = plt.subplots(figsize= (10,8))

ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending= False).index],
    labels = pop_pivot.iloc[-1].sort_values(ascending= False).index,
    
)
ax.legend(loc = "upper left")
ax.set_title("Growth of Population over Time")
ax.set_xlabel("Year")
ax.set_xticks(gdp_pivot.index[::5])
ax.set_ylabel("Population (Billions)")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False);
```

```python
# Create bubble chart with life expectancy for each country and GDP per capita in Year 2014
import matplotlib.ticker as mticker

fig, ax = plt.subplots(figsize = (10,8))

minsize = min(hdi_world_2014["Population (M)"])
maxsize = max(hdi_world_2014["Population (M)"])

sns.scatterplot(
    x = "Life expectancy at birth (years)", y = "GDP per capita (USD)", data = hdi_world_2014, 
    size = "Population (M)", sizes= (minsize, maxsize), hue = "Region",
    ax=ax
).set(title = "Life Expectancy vs Countries GDP", yscale="log")

ax.legend(loc = "upper left")

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
sns.despine();
```

```python
# bar chart of avg HDI by region. 
wb_hdi_by_region
```

```python
fig, ax = plt.subplots(figsize = (8,6))

sns.barplot(data = wb_hdi_by_region, x= "Region", y = "hdi_2014"
           ).set(ylabel = "Human Development Index (HDI)", title = "HDI by Region")
plt.xticks(rotation = 90)
sns.despine();
```

```python
import matplotlib.gridspec as gridspec
#import matplotlib as mpl
import matplotlib.pyplot as plt

   

fig = plt.figure(figsize= (9,14))

gs = gridspec.GridSpec(nrows=12, ncols = 6)

# mpl.rcParams['font.family'] = "Lato"

stack1_list = ['#66C2A5', '#FC8D62', '#8DA0CB', '#E78AC3', '#A6D854', '#FFD92F', '#B3B3B3']
stack2_list = ['#66C2A5', '#A6D854', '#B3B3B3', '#FC8D62', '#E78AC3', '#8DA0CB', '#FFD92F']
bar_list    = ['#8DA0CB', '#fC8D62', '#E78AC3', '#FFD92F', '#66C2A5', '#A6D854', '#B3B3B3']
bubble_list = ['#A6D854', '#FC8D62', '#FFD92F', '#66C2A5', '#B3B3B3', '#E78AC3', '#8DA0CB']

fig.suptitle(
    "Global Economic Report: 1960-2018             ",
    fontsize = 16,
    fontweight = "bold",
    x=.65,
    y=.93
)

ax1 = fig.add_subplot(gs[0:4, 0:3])
# ax1.plot(
#     gdp_pivot.index,
#     [gdp_pivot[region]/1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending= False).index],
#     labels = gdp_pivot.iloc[-1].sort_values(ascending= False).index,
#     colors = stack1_list
# )
# ax1.set_title("GDP exponential over Time")
# ax1.set_ylabel("GDP (Trillions) ")

# Loop through each region to add separate line plots for each region
for region, color in zip(gdp_pivot.iloc[-1].sort_values(ascending=False).index, stack1_list):
    ax1.plot(
        gdp_pivot.index,
        gdp_pivot[region] / 1_000_000_000_000,  # Divide by 1 trillion to convert to trillions
        label=gdp_pivot.iloc[-1].sort_values(ascending= False).index,                           # Each region has a separate label
        color=color                            # Set color for each line
    )

#ax1.legend()  # Add legend for the labels
ax1.set_title("GDP exponential over Time")
ax1.set_ylabel("GDP (Trillions)")
ax1.grid(True, which='both', linestyle='--', linewidth=0.5)
#ax1.set_xlabel("Year")
#ax1.set_xticks(gdp_pivot.index[::5])
#ax1.spines['right'].set_visible(False)
#ax1.spines['top'].set_visible(False)
#ax1.grid(True, which='both', linestyle='--', linewidth=0.5) 


fig.text(.9, .61,
         '''
         Economic growth in the 20th and early
         21st centuries has been rapid, outpacing
         population increases. In 1960, Europe and
         North America generated 75% of global GDP
         with only 20% of the world's population.

         By 2018, East Asia surpassed in GDP, driven
         by Japan, South Korea and China.

         In coming decades, South Asia and Sub-Saharan
         Africa will lead global economic growth.
         '''
         , fontsize = 10.5)


fig.text(.9, .45,
         '''
         The world's wealthiest nations enjoy long life 
         spans, yet represent small share of population.

         The big green circle represents India, and
         other big circle represents China.
         '''
         , fontsize = 10.5)

fig.text(.9, .15,
         '''
         HDI, Human Development Index, aims to measure
         the overall standard of living in countries.
         It considers factors such as life expectancy,
         GDP per capita, and educational attainment.

         Economic growth in the developing world during
         21st century may help other regions catch up
         to North America and Europe.


         Access to electricity enables significant productivity
         gains. Developing nations need to invest in energy
         production to foster growth.
         '''
         , fontsize = 10.5)


ax2 = fig.add_subplot(gs[0:4, 3:6])
ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending= False).index],
    labels = pop_pivot.iloc[-1].sort_values(ascending= False).index,
    colors = stack2_list
)
ax2.legend(bbox_to_anchor=(1.95, 1))
ax2.set_title("Population grew from 2 billion to 7.5 billion")
ax2.set_ylabel("Population (Billions)")
#ax.set_xlabel("Year")
#ax.set_xticks(gdp_pivot.index[::5])
#ax2.grid(True, which='both', linestyle='--', linewidth=0.5) 


ax3 = fig.add_subplot(gs[4:8, :])
minsize = min(hdi_world_2014["Population (M)"])
maxsize = max(hdi_world_2014["Population (M)"])
sns.scatterplot(
    x = "Life expectancy at birth (years)", y = "GDP per capita (USD)", data = hdi_world_2014, 
    size = "Population (M)", sizes= (minsize, maxsize), hue = "Region", 
    palette = bubble_list, ax=ax3
).set(title = "Life Expectancy increases as Countries get Richer", yscale="log")
ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())
handles, labels = ax3.get_legend_handles_labels()
entries_to_skip = len(hdi_world_2014["Region"].unique())+ 1
ax3.legend(handles[entries_to_skip:], labels[entries_to_skip:], bbox_to_anchor=(.75, .9), borderaxespad=0., frameon= False, ncols =6, fontsize = 9)
#ax3.legend(loc = "upper left")
ax3.set_xlabel("Life expectancy (Years)", fontsize=10)


ax4 = fig.add_subplot(gs[8:12, 0:3])
ax4=sns.barplot(data = wb_hdi_by_region, x= "Region", y = "hdi_2014", hue = bar_list, ax = ax4, legend = False
           ).set(ylabel = "Human Development Index (HDI)", xlabel="", title = "HDI by Region")
plt.xticks(rotation = 90)


ax5 = fig.add_subplot(gs[8:12, 3:6])
sns.scatterplot(data = hdi_world_2014.query("Country != 'Iceland'"), 
                x = "Electric power consumption (kWh per capita)", y = "GDP per capita (USD)", 
                hue = "hdi_2014", palette = "coolwarm_r", alpha = .7, ax= ax5
               ).set(title = "More Electricity, more development")
ax5.legend(bbox_to_anchor=(1, .8))

fig.subplots_adjust(wspace=4, hspace=4)

fig.savefig('Economic Report Project.png', bbox_inches='tight', pad_inches=0.2)
```
